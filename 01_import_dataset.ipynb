## Selections of reciproque match at the first wave meet

```python
wave1 = dataset[dataset['wave'] == 1]
wave1
```

## Keep Only reciproque match

```python
# Keep Only the Match for Match 
match = wave1[wave1['match'] == 1]

# Separate gender
female = match[match['gender'] == 0]
male = match[match['gender'] == 1]

# iid: unique subject number
# pid: partner’s iid number

# Merge reciproque match
match_for_match = pd.merge(male, female, left_on='pid', right_on='iid')

# Drop duplicate (in case of)
match_for_match.drop_duplicates(subset=None, keep='first', inplace=True, ignore_index=False)

match_for_match
```
